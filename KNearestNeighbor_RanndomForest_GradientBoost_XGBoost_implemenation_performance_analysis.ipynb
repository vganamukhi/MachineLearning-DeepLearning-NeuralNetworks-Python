```python
from sklearn.model_selection import train_test_split
Xcols = list(Model_data)
Xcols.remove("y")

X = Model_data[Xcols]
Y = Model_data["y"]

# split training and testing
test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=8)
```

```python
#KNN model

from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier  
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix 
import random


accuracy_test_knn = []




for i in range(5,17,2):
    print("neighbours =",i)
    
    vars = random.sample(Xcols,50)
    x_train_knn = X_train.loc[:,vars]
    x_test_knn = X_test.loc[:,vars]
    
    knn = KNeighborsClassifier(n_neighbors = i, algorithm = 'kd_tree', p = 2)
    knn.fit(x_train_knn, y_train) 
    knn_pred=knn.predict(x_test_knn)
    print("test:")
    print(accuracy_score(knn_pred, y_test))
    accuracy_test_knn.append(accuracy_score(knn_pred,y_test))
```

```python
accuracy_test_knn
```

```python
#Grid Search
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.model_selection import StratifiedKFold as skf
from xgboost import XGBClassifier

#xg bosst model

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=8)
accuracy_test_xg = []

# A parameter grid for XGBoost
max_depth = [9, 10,11]
learning_rate = [0.1, 0.01, 0.001]
subsample = [0.6, 0.8, 1.0]

# n_estimators = 100
for i in max_depth:
    for j in learning_rate:
        for k in subsample:
            print(i,j,k)
            model = XGBClassifier(n_estimators =100, learning_rate=j, subsample = k, max_depth=i)
            model.fit(X_train, y_train)
            pred = model.predict(X_test)
            print(accuracy_score(pred, y_test))
            accuracy_test_xg.append(accuracy_score(pred,y_test))
            
        

# n_estimators = 500

for i in max_depth:
    for j in learning_rate:
        for k in subsample:
            print(i,j,k)
            model = XGBClassifier(n_estimators =500, learning_rate=j, subsample = k, max_depth=i)
            model.fit(X_train, y_train)
            pred = model.predict(X_test)
            print(accuracy_score(pred, y_test))
            accuracy_test_xg.append(accuracy_score(pred,y_test))


```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier

#Random forest model

accuracy_test_rf = []

# n_estimators = 100
model = RandomForestClassifier(n_estimators =100, min_samples_split = 5, min_samples_leaf = 5, random_state=8)
model.fit(X_train, y_train)
pred = model.predict(X_test)
print(accuracy_score(pred, y_test))
accuracy_test_rf.append(accuracy_score(pred,y_test))
            
# n_estimators = 500
model = RandomForestClassifier(n_estimators =500, min_samples_split = 5, min_samples_leaf = 5, random_state=8)
model.fit(X_train, y_train)
pred = model.predict(X_test)
print(accuracy_score(pred, y_test))
accuracy_test_rf.append(accuracy_score(pred,y_test))
```

```python
# A parameter grid for GBoost
#Gradient boost

accuracy_test_gb = []
f1_score_test_gb = []


learning_rate = [0.1, 0.01, 0.001]
subsample = [0.6, 0.8, 1.0]

# n_estimators = 100
for j in learning_rate:
    for k in subsample:
        print(j,k)
        model = GradientBoostingClassifier(learning_rate=j, n_estimators =100, subsample = k, 
                                           min_samples_split = 5, min_samples_leaf = 5, random_state=8)
        model.fit(X_train, y_train)
        pred = model.predict(X_test)
        print(accuracy_score(pred, y_test))
        accuracy_test_gb.append(accuracy_score(pred,y_test))

# n_estimators = 500
for j in learning_rate:
    for k in subsample:
        print(j,k)
        model = GradientBoostingClassifier(learning_rate=j, n_estimators =500, subsample = k, 
                                           min_samples_split = 5, min_samples_leaf = 5, random_state=8)
        model.fit(X_train, y_train)
        pred = model.predict(X_test)
        print(accuracy_score(pred, y_test))
        accuracy_test_gb.append(accuracy_score(pred,y_test))
```
