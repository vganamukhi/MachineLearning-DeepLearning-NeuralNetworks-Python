```python
from  sklearn.datasets import make_classification 
X,y = make_classification (n_samples=10000, n_features=18, n_informative=8, n_redundant=2, 
          n_repeated=0, n_classes=2, n_clusters_per_class=4, weights =None, flip_y=0.2, 
          class_sep=0.1, hypercube=False, shift=1, scale=1, shuffle=True, random_state = 92846)
```

```python
scores_layer1 = []
history = []
def create_model_layer1(batch_, activation_, nodes_):
    print("layer1:", nodes_, batch_, activation_)
    model = Sequential()
    model.add(Dense(nodes_, batch_size = batch_, activation = activation_))
    model.add(Dense(1, activation='sigmoid'))
    # Compile model
    model.compile(loss='binary_crossentropy', optimizer='RMSprop', metrics=['accuracy'])
    history = model.fit(X, y, epochs=200, batch_size=batch_)
    print(history.history['accuracy'])
    history = history.history['accuracy']
    print(type(history))
    scores_ = model.evaluate(X, y)
    print("%s: %.2f%%" % (model.metrics_names[1], scores_[1]*100))
    #print(model.metrics_names)
    print(scores_[1]*100)
    scores_layer1.append(scores_[1]*100)
    
    cols = []
    for i in range(1,201):
        cols.append("E_"+str(i))
        
    print(cols)
        
    acc_df = pd.DataFrame(data = history)
    acc_df = acc_df.T
    acc_df.columns = cols
    new = pd.DataFrame()
    new.loc[0,'Batch_size'] = batch_
    new.loc[0,'Activation_function'] = activation_
    new.loc[0,'Nodes'] = nodes_
    new.loc[0,'Layer'] = 1
    
    new = pd.concat([new, acc_df], axis = 1)
    
    print(new)
    
    return new

scores_layer2 = []
def create_model_layer2(batch_, activation_, nodes_):
    print("layer2:", nodes_, batch_, activation_)
    model = Sequential()
    model.add(Dense(nodes_, batch_size = batch_, activation = activation_))
    model.add(Dense(nodes_, batch_size = batch_, activation = activation_))
    model.add(Dense(1, activation='sigmoid'))
    # Compile model
    model.compile(loss='binary_crossentropy', optimizer='RMSprop', metrics=['accuracy'])
    history = model.fit(X, y, epochs=200, batch_size=batch_)
    print(history.history['accuracy'])
    history = history.history['accuracy']
    scores_ = model.evaluate(X, y)
    print("%s: %.2f%%" % (model.metrics_names[1], scores_[1]*100))
    #print(model.metrics_names)
    print(scores_[1]*100)
    scores_layer2.append(scores_[1]*100)
    
    cols = []
    for i in range(1,201):
        cols.append("E_"+str(i))
        
    acc_df = pd.DataFrame(data = history)
    acc_df = acc_df.T
    acc_df.columns = cols
    
    new = pd.DataFrame()
    new.loc[0,'Batch_size'] = batch_
    new.loc[0,'Activation_function'] = activation_
    new.loc[0,'Nodes'] = nodes_
    new.loc[0,'Layer'] = 2
    
    new = pd.concat([new, acc_df], axis = 1)
    print(new)
    
    return new

leaky_scores_layer1 = []
def leaky_relu_model_layer1(batch_, nodes_):
    print("Laeky Relu layer1:", nodes_, batch_)
    model = Sequential()
    model.add(Dense(nodes_, batch_size = batch_))
    model.add(LeakyReLU(alpha=0.05))
    model.add(Dense(1, activation='sigmoid'))
    # Compile model
    model.compile(loss='binary_crossentropy', optimizer='RMSprop', metrics=['accuracy'])
    history = model.fit(X, y, epochs=200, batch_size=batch_)
    print(history.history['accuracy'])
    history = history.history['accuracy']
    print(type(history))
    scores_ = model.evaluate(X, y)
    print("%s: %.2f%%" % (model.metrics_names[1], scores_[1]*100))
    #print(model.metrics_names)
    print(scores_[1]*100)
    leaky_scores_layer1.append(scores_[1]*100)
    
    cols = []
    for i in range(1,201):
        cols.append("E_"+str(i))
        
    print(cols)
        
    acc_df = pd.DataFrame(data = history)
    acc_df = acc_df.T
    acc_df.columns = cols
    new = pd.DataFrame()
    new.loc[0,'Batch_size'] = batch_
    new.loc[0,'Activation_function'] = 'Leaky_ReLU'
    new.loc[0,'Nodes'] = nodes_
    new.loc[0,'Layer'] = 1
    
    new = pd.concat([new, acc_df], axis = 1)
    
    print(new)
    
    return new


leaky_scores_layer2 = []    
def leaky_relu_model_layer2(batch_, nodes_):
    print(" Laeky Relu model layer2:", nodes_, batch_, activation_)
    model = Sequential()
    model.add(Dense(nodes_, batch_size = batch_))
    model.add(LeakyReLU(alpha=0.05))
    model.add(Dense(nodes_, batch_size = batch_))
    model.add(LeakyReLU(alpha=0.05))
    model.add(Dense(1, activation='sigmoid'))
    # Compile model
    model.compile(loss='binary_crossentropy', optimizer='RMSprop', metrics=['accuracy'])
    history = model.fit(X, y, epochs=200, batch_size=batch_)
    print(history.history['accuracy'])
    history = history.history['accuracy']
    scores_ = model.evaluate(X, y)
    print("%s: %.2f%%" % (model.metrics_names[1], scores_[1]*100))
    #print(model.metrics_names)
    print(scores_[1]*100)
    leaky_scores_layer2.append(scores_[1]*100)
    
    cols = []
    for i in range(1,201):
        cols.append("E_"+str(i))
        
    acc_df = pd.DataFrame(data = history)
    acc_df = acc_df.T
    acc_df.columns = cols
    
    new = pd.DataFrame()
    new.loc[0,'Batch_size'] = batch_
    new.loc[0,'Activation_function'] = 'Leaky_ReLU'
    new.loc[0,'Nodes'] = nodes_
    new.loc[0,'Layer'] = 2
    
    new = pd.concat([new, acc_df], axis = 1)
    print(new)
    
    return new

batch_size = [8,16,32]
#epochs = [10, 50, 100]
activation = ['sigmoid', 'relu']
neurons = [10,20,50]

results = pd.DataFrame()

for batch_ in batch_size:
    for nodes_ in neurons:
        for activation_ in activation:
            acc_df1 = create_model_layer1(batch_, activation_, nodes_)
            acc_df2 = create_model_layer2(batch_, activation_, nodes_)
            results = pd.concat([results, acc_df1], axis = 0)
            results = pd.concat([results, acc_df2], axis = 0)
            
        leaky_acc_df1 = leaky_relu_model_layer1(batch_, nodes_)
        leaky_acc_df2 = leaky_relu_model_layer2(batch_, nodes_)
        results = pd.concat([results, leaky_acc_df1], axis = 0)
        results = pd.concat([results, leaky_acc_df2], axis = 0)
            
    
```
